```python
class SLinkedList:
    def __init__(self):
        self.head = None

    def addAtHead(self, data):
        node = Node(data)
        node.next = self.head
        self.head = node
```

```python
slist = SLinkedList()
slist.addAtHead(1)
slist.addAtHead(2)

printNodes(slist.head)
```
